```python
import baltic as bt
import pandas as pd
import numpy as np
import json
import os
import matplotlib.pyplot as plt
import itertools
from collections import defaultdict
import statistics
import numpy as np
import re
import random
import math
import scipy.stats as stats
from collections import Counter
from scipy.stats import gaussian_kde
import seaborn as sns
```

```python
clades = ['na_avian', 'eurasian_avian', 'swine']
```

```python
host_trees = {}

for clade in clades:
    
    path = f'hosts/{clade}/summary_baltic.nwk'
    mytree= bt.loadNewick(path, absoluteTime= False)

    
    host_trees[clade] = mytree
```

```python
print(host_trees)
```

```python
for clade in clades:
    
    mytree = host_trees[clade]

    reassortments = []

    for k in mytree.Objects:

        if k.traits["is_reassorted"]:

            reassortments.append(k.traits["rea"])

    nodes = list(mytree.Objects)

    for k in nodes:
        k.traits["is_reassorted"] = 0
        k.traits.pop("rea", None)

    for i in range(1000):

        shuffled_nodes = random.sample(nodes, len(reassortments))

        for node, rea_label in zip(shuffled_nodes, reassortments):
            node.traits["is_reassorted"] = 1
            node.traits["rea"] = rea_label

        # writing to new file
        tree = mytree.toString()
        with open(f"hosts/{clade}/trees/randomized_tree_{i}.nwk", "w") as f:
            f.write(tree)

        # resetting nodes that were assigned as reassorted at this iteration
        for node in shuffled_nodes:
            node.traits["is_reassorted"] = 0
            del node.traits["rea"]
```
